## Question:
Please rewrite the code such that the response to a yes/no question would be boolean (`True` or `False`)

## preprocessing in order to deal with none returning from GPT

```python
df["category_names"]
```

```python
df['for_kids']
```

```python
# Returns TRUE if it is a category of games without an element of risk
def is_kids_game(category_string):
  kids_friendly_categories = ["GAME_PUZZLE", "GAME_CASUAL", "GAME_WORD", "GAME_ADVENTURE", "GAME_BOARD"]
  categories = category_string.split(",")
  for category in categories:
    if category in kids_friendly_categories:
      return True
  return False
```

```python
# validation check
is_kids_game("GAME_PUZZLE,GAME")
```

```python
# Cleans the unnecessary characters, and converts to boolean
def convert_yes_no_to_boolean(df):
  for index, row in df.iterrows():
    value = row["for_kids"]
    # for the None cases - check by category
    if value == None:
      df.loc[index, "response"] = is_kids_game(row["category_names"])
      continue
    if value.lower().strip() == 'yes':
      df.loc[index, "response"] = True
    elif value.lower().strip() == 'no':
      df.loc[index, "response"] = False
```

```python
# activation
convert_yes_no_to_boolean(df)
```
